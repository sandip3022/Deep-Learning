Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Importing Datasets

### Splitting data into Depedent varibles and Indepedent variables

In our dataset **Indepedent Varibles** are:
* CreditScore
* Geography
* Gender
* Age
* Tenure
* Balance
* NumOfProducts
* HasCrCard

On the basis of above feature we are going to predict `Exited` i.e. **Depedent Variable**

```python
x = data.iloc[:,:-1]
x
```

```python
y = data.iloc[:,-1]
y
```

### Creating Dummy Variables

```python
geo = pd.get_dummies(x["Geography"],drop_first=True)
gender = pd.get_dummies(x["Gender"],drop_first=True)
```

```python
# Concatenate the Dataframe
x = pd.concat([x,geo,gender],axis=1)
x
```

```python
# Drop Unnecessary columns
x=x.drop(['Geography','Gender'],axis=1)
```

```python
x
```

### Splitting the dataset into the training and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train),len(y_train),len(x_test),len(y_test)
```

# Perform Hyperparameter Optimization

```python
# !pip install scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Embedding,Flatten,LeakyReLU,BatchNormalization, Dropout
from keras.activations import relu, sigmoid
```

```python
def create_model(layers, activation):
        model= Sequential()
        for i, nodes in enumerate(layers):
            if i==0:
                model.add(Dense(nodes, input_dim=x_train.shape[1]))
                model.add(Activation(activation))
            else:
                model.add(Dense(nodes))
                model.add(Activation(activation))
        model.add(Dense(1))

        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model
```

```python
model= KerasClassifier(model=create_model, verbose=0, activation='relu', layers=20)
```

```python
layers=[[20],[40,20], [45, 30, 15]]
activations = ['sigmoid','relu']
param_grid = dict(layers=layers, activation=activations, batch_size=[128, 256], epochs=[30])
grid = GridSearchCV(estimator=model, param_grid=param_grid)
```

```python
grid_result = grid.fit(x_train,y_train)
```

```python
[grid_result.best_score_,grid_result.best_params_]
```

```python
pred_y = grid.predict(x_test)
```

```python
y_pred = (pred_y > 0.5)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
```
