## Data Augmentation using Python and Keras

### Importing Required Library

```python
from keras.preprocessing.image import ImageDataGenerator # For Data Augmentation
from keras.preprocessing.image import load_img
from keras.preprocessing.image import array_to_img # for to convert array to image
from keras.preprocessing.image import img_to_array # For to convert image to array
```

```python
datagen = ImageDataGenerator( rotation_range=40,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             fill_mode="nearest")
```

```python
# Lets load Cat
cat = load_img("Training Data/cat.png")
cat
```

```python
# Convert Image to array
x = img_to_array(cat)
x.shape
```

```python
x[:5,:5]
```

**Reshape array**

```python
x = x.reshape((1,)+x.shape)
x.shape
```

The `.flow()` command generate batches of randomly transformed image 
and save the result in `preview` directory.


```python
i = 1
for batch in datagen.flow(x,batch_size=1,save_to_dir="Generated Images",save_format="jpeg"):
    i+=1
    if i > 30:
        break    
```
