## Import Libraries

```python
from tensorflow.keras.preprocessing.text import one_hot
import numpy as np
```

## Senteces

```python
sen = [ 'the glass of milk',
       'the glass of juice',
       'the cup of tea',
       'I am a good boy',
       'I am good developer',
       'understand the meaning of words',
       'your videos are good',
       'the new good videos are easy to understand']
```

```python
sen
```

```python
# Vocabulary size
voc_size = 10000
```

```python
onehot_represent = [one_hot(words,voc_size) for words in sen]
```

```python
onehot_represent
```

## Word Embedding Representation

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
```

```python
sen_length = 8
embeded_doc = pad_sequences(onehot_represent,padding='pre',maxlen=sen_length)
```

```python
print(embeded_doc)
```

```python
dim = 15
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length=sen_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embeded_doc)
```
